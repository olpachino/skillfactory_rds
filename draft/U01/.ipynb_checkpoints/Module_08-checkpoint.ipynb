```python
sample = pd.read_csv("sample.csv")
sample2 = sample
sample2['City'] = sample2.City.astype(str)
sample2['City'] = sample2.City.apply(lambda s: s.lower())
sample2
```

```python
def profession_code(s):
    if s == 'Рабочий':
        return 0
    elif s == 'Менеджер':
        return 1
    else:
        return 2

sample = pd.read_csv("sample.csv")
sample2 = sample
sample2['Profession'] = sample2.Profession.apply(profession_code)
sample2
```

```python
sample = pd.read_csv("sample.csv")


def age_category(age):
    if age < 23:
        return 'молодой'
    elif (23 < age) and (age < 35):
        return 'средний'
    else:
        return 'зрелый'


sample['Age_category'] = sample.Age.apply(age_category)
sample
```

```python
log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id','time','bet','win']


def correct_user(user_id):
    if 'error' in user_id:
        return ''
    else:
        user = user_id.split(' ')
        return user[-1]


log['user_id'] = log.user_id.apply(correct_user)
log
```

```python
log = pd.read_csv("log.csv",header=None) 
log.columns = ['user_id','time','bet','win']


def correct_time(t):
    if pd.notnull(t):
        list_t = list(t)[1:]
        t = ''.join(list_t)
        return t
    else:
        return t


log['time'] = log.time.apply(correct_time)
log
```
