```python
import pandas as pd
log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id', 'time', 'bet', 'win']
log = log[~log.user_id.str.contains("error", na=False)]


def correct_user(user_id):
    user = user_id.split()
    return user[-1]


log['user_id'] = log.user_id.apply(correct_user)


def correct_time(time):
    if pd.notnull(time):
        list_time = list(time)[1:]
        time = ''.join(list_time)
        return time
    else:
        return time


log['time'] = log.time.apply(correct_time)
```
