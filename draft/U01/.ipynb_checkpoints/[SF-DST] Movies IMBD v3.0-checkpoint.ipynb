# 11. Какого жанра фильмов больше всего?
Варианты ответов:
1. Action
2. Adventure
3. Drama
4. Comedy
5. Thriller

```python
answer_ls.append(3)
```

```python
import collections

def list_uniq(df_col):
    df = df_col.str.split('|')
    count_df = collections.Counter()
    for list_df in df:
        for pos in list_df:
            count_df[pos] += 1
    return count_df

genre = list_uniq(data.genres)
genre.most_common()
```

# 15. Кто из режиссеров принес больше всего прибыли?
Варианты ответов:
1. Steven Spielberg
2. Christopher Nolan
3. David Yates
4. James Cameron
5. Peter Jackson


```python
answer_ls.append(5)
```

```python
directors = list_uniq(data.director)
df_dir = pd.DataFrame(list(directors), columns = ['name'])

def total_prof(col_count, df, df_col):
    total_profit = []
    for name in list(col_count):
        total_profit.append(df[df_col.str.contains(name, na=False)].profit.sum())
    return total_profit

df_dir['total_profit'] = total_prof(directors, data, data.director)
df_dir[df_dir['total_profit'] == df_dir['total_profit'].max()]
```

# 31. Названия фильмов какой студии в среднем самые длинные по количеству символов?
Варианты ответа:
1. Universal Pictures (Universal)
2. Warner Bros
3. Jim Henson Company, The
4. Paramount Pictures
5. Four By Two Productions

```python
answer_ls.append(5)
```

```python
comp = list_uniq(data.production_companies)
df_comp = pd.DataFrame(list(comp), columns = ['name'])

df_titles = []
for title in data.original_title:
    df_titles.append(len(title))

data['len_titles'] = df_titles

mean_len = []
for name in list(comp):
    mean_len.append(data[data['production_companies'].str.contains(name, na=False)].len_titles.mean())

df_comp['mean_len'] = mean_len
df_comp[df_comp['mean_len'] == df_comp['mean_len'].max()]
```
