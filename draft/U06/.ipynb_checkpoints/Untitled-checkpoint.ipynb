```python
import pandas as pd

import math

import requests
from bs4 import BeautifulSoup

from datetime import datetime
```

```python
def get_html(url):
    r = requests.get(url).encoding('utf-8')
    return r.text  # возвращает HTML-код страницы 
```

```python
''' Функция не отрабатывает должным образом, так как основная масса ссылок на бренды скрыта кнопкой.
    Функция выводит только 11 самых популярных брендов по объявлениям.
    Решить данную проблему я не смогла.'''

def get_brend_auto(html):
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find('div', class_='ListingPopularMMM-module__items')
    
    divs = table.find_all('div', class_='ListingPopularMMM-module__item')
    brends = {}
    for div in divs:
        try:
            brand = div.find('a', class_='Link ListingPopularMMM-module__itemName').text.strip()
        except:
            brand = ''
        
        try:
            count = div.find('div', class_='ListingPopularMMM-module__itemCount').text.strip()
        except:
            count = ''
            
        try:
            link = div.find('a').get('href')
        except:
            link = ''
        
        brends[brend] = [count, link]
    
    return brends

url = 'https://auto.ru/moskva/cars/used/'
brends = get_brend_auto( get_html(url) )
```

```python
def get_links_brend(brends):
    all_start = datetime.now()
    all_links = []
    
    for brend in brends:
        url_brend = used_brends[brend][1]
        time_start = datetime.now()
        for page in range( math.ceil(brends[brend][0] / 37) ):
            url_page = url_brend + '?page=' + str(page)
            html_page = get_html(url_page)
            soup_page = BeautifulSoup(html_page, 'html.parser')
            a_page = soup_page.find_all('a', class_='Link ListingItemTitle-module__link')
            
            for a in a_page:
                link = a.get('href')
                all_links.append(link)
        time_end = datetime.now()
        print( brend, 'завершен', str(time_end - time_start) )
    
    all_end = datetime.now()
    print('Весь процесс занял', str(all_end-all_start))
    return all_links
```

```python
all_links = get_links_brend(used_brends)
```
