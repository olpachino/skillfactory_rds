```python
def get_html(url):
    r = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    r.encoding = 'utf-8'
    html = r.text
    
    return html  # возвращает HTML-код страницы 
```

```python
def get_links_brend(brends):
    all_start = datetime.now()
    all_links = []
    
    for brend in brends:
        url_brend = used_brends[brend][1]
        time_start = datetime.now()
        for page in range( math.ceil(brends[brend][0] / 37) ):
            url_page = url_brend + '?page=' + str(page)
            html_page = get_html(url_page)
            soup_page = BeautifulSoup(html_page, 'html.parser')
            a_page = soup_page.find_all('a', class_='Link ListingItemTitle-module__link')
            
            for a in a_page:
                link = a.get('href')
                all_links.append(link)
        time_end = datetime.now()
        print( brend, 'завершен', str(time_end - time_start) )
    
    all_end = datetime.now()
    print('Весь процесс занял', str(all_end-all_start))
    return all_links
```

```python
all_links = get_links_brend(used_brends)
```

```python
def get_tech_char_model(url):
    
    page_model = get_html(url)
    model_soup = BeautifulSoup(page_model, 'html.parser')
    
 
    try:
        engine_displacement = model_soup.find('div', class_='catalog__details-main clearfix'
                                             ).find_all('div', class_='catalog__column catalog__column_half'
                                                       )[2].find_all('dd', class_='list-values__value')[0].text.strip()
    except:
        engine_displacement = ''    
    try:
        engine_power = model_soup.find('div', class_='catalog__details-main clearfix'
                                      ).find_all('div', class_='catalog__column catalog__column_half'
                                                )[2].find_all('dd', class_='list-values__value')[1].text.strip()
    except:
        engine_power = '' 
    try:
        vehicle_transmission = model_soup.find('div', class_='catalog__details-main clearfix'
                                              ).find_all('div', class_='catalog__column catalog__column_half'
                                                        )[2].find_all('dd', class_='list-values__value')[2].text.strip()
    except:
        vehicle_transmission = ''
    try:
        fuel_type = model_soup.find('div', class_='catalog__details-main clearfix'
                                   ).find_all('div', class_='catalog__column catalog__column_half'
                                             )[2].find_all('dd', class_='list-values__value')[3].text.strip()
    except:
        fuel_type = ''
    try:
        drive_unit = model_soup.find('div', class_='catalog__details-main clearfix'
                                    ).find_all('div', class_='catalog__column catalog__column_half'
                                              )[3].find_all('dd', class_='list-values__value')[1].text.strip()
    except:
        drive_unit = ''
    try:
        vendor = model_soup.find_all('dd', class_='list-values__value')[8].text.strip()
    except:
        vendor = ''
    try:
        number_of_doors = model_soup.find_all('dd', class_='list-values__value')[10].text.strip()
    except:
        number_of_doors = ''
    
    dict_tech = {'engine_displacement':engine_displacement,
                 'engine_power':engine_power, 'vehicle_transmissione':vehicle_transmission, 
                 'fuel_type':fuel_type, 'drive_unit':drive_unit, 'vendor':vendor, 
                 'number_of_doors':number_of_doors}
    
    return dict_tech
```

```python
def get_data(url):
    
    page_car = get_html(url)
    car_soup = BeautifulSoup(page_car, 'html.parser')
    
    try:
        brand = car_soup.find('div', class_='CardBreadcrumbs'
                             ).find_all('div', class_='CardBreadcrumbs__item'
                                       )[1].find('a', class_='Link Link_color_gray CardBreadcrumbs__itemText'
                                                ).text.strip()
    except:
        brand = ''
    try:
        model_name = car_soup.find('div', class_='CardBreadcrumbs'
                             ).find_all('div', class_='CardBreadcrumbs__item'
                                       )[2].find('a', class_='Link Link_color_gray CardBreadcrumbs__itemText'
                                                ).text.strip()
    except:
        model_name = ''
    try:
        generation = car_soup.find('div', class_='CardBreadcrumbs'
                             ).find_all('div', class_='CardBreadcrumbs__item'
                                       )[3].find('a', class_='Link Link_color_gray CardBreadcrumbs__itemText'
                                                ).text.strip()
    except:
        generation = ''
    
    try:
        price = car_soup.find('span', class_='OfferPriceCaption__price').text.strip()
    except:
        price = ''
    try:
        sell_id = car_soup.find('div', class_='gRe_0imRLXMjOwJ0pjt6N__info-item',
                                title='Идентификатор объявления').text.strip()
    except:
        sell_id = ''
    try:
        production_date = car_soup.find('li', class_='CardInfoRow CardInfoRow_year'
                                       ).find('a',class_='Link Link_color_black').text.strip()
    except:
        production_date = ''
    try:
        mileage = car_soup.find('li', class_='CardInfoRow CardInfoRow_kmAge'
                               ).find_all('span', class_='CardInfoRow__cell')[1].text.strip()
    except:
        mileage = ''
    try:
        body_type = car_soup.find('li', class_='CardInfoRow CardInfoRow_bodytype'
                                 ).find_all('span', class_='CardInfoRow__cell')[1].text.strip()
    except:
        body_type = ''
    try:
        color = car_soup.find('li', class_='CardInfoRow CardInfoRow_color'
                                 ).find_all('span', class_='CardInfoRow__cell')[1].text.strip()
    except:
        color = ''
    try:
        steering_position = car_soup.find('li', class_='CardInfoRow CardInfoRow_wheel'
                                         ).find_all('span', class_='CardInfoRow__cell')[1].text.strip()
    except:
        steering_position = ''
    try:
        condition = car_soup.find('li', class_='CardInfoRow CardInfoRow_state'
                                 ).find_all('span', class_='CardInfoRow__cell')[1].text.strip()
    except:
        condition = ''
    try:
        owners = car_soup.find('li', class_='CardInfoRow CardInfoRow_ownersCount'
                              ).find_all('span', class_='CardInfoRow__cell')[1].text.strip()
    except:
        owners = ''
    try:
        possession = car_soup.find('li', class_='CardInfoRow CardInfoRow_owningTime'
                                  ).find_all('span', class_='CardInfoRow__cell')[1].text.strip()
    except:
        possession = ''
    try:
        PTS = car_soup.find('li', class_='CardInfoRow CardInfoRow_pts'
                           ).find_all('span', class_='CardInfoRow__cell')[1].text.strip()
    except:
        PTS = ''
    try:
        customs = car_soup.find('li', class_='CardInfoRow CardInfoRow_customs'
                               ).find_all('span', class_='CardInfoRow__cell')[1].text.strip()
    except:
        customs = ''
    try:
        description = car_soup.find('div', class_='CardDescription__textInner').text.strip()
    except:
        description = ''
    
    try:
        url_tech_model = car_soup.find('a',
                                       class_='Link SpoilerLink CardCatalogLink SpoilerLink_type_default').get('href')
    except:
        url_tech_model = ''
    
    dict_tech = get_tech_char_model(url_tech_model)

    dict_piece = {'brand':brand, 'model_name':model_name, 'generation':generation,
                  'url_car':url, 'price':price, 'sell_id':sell_id, 'production_date':production_date,
                 'mileage':mileage, 'body_type':body_type, 'color':color, 'steering_position':steering_position,
                 'condition':condition, 'owners':owners, 'possession':possession, 'PTS':PTS,
                 'customs':customs, 'description':description}
    
    dict_data = {**dict_piece, **dict_tech}
    print('Парсинг страницы завершен')
    sleep(5)
    return dict_data
```
