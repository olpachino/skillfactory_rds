```python
# Загрузка библиотек
import numpy as np # для работы с массивами
import pandas as pd # для работы DataFrame 
import seaborn as sns # библиотека для визуализации статистических данных
import matplotlib.pyplot as plt # для построения графиков

%matplotlib inline
```

```python
df = pd.read_csv('Admission_Predict_Ver1.1.csv')
```

В датасете Admission_Predict_Ver1.1.csv находятся данные об абитуриентах и вероятности приёма их в магистратуру.

Целевой переменной является вероятность приёма в университет — Chance of Admit (ranging from 0 to 1 )

Признаки: 

1. GRE Scores ( out of 340 ) — баллы по GRE тестированию
2. TOEFL Scores ( out of 120 ) — баллы по экзамену TOEFL
3. University Rating ( out of 5 ) — рейтинг университета 
4. Statement of Purpose Strength ( out of 5 ) — весомость мотивационного письма
5. Letter of Recommendation Strength ( out of 5 ) — весомость рекомендационных писем
6. Undergraduate GPA ( out of 10 ) — средний балл в бакалавриате
7. Research Experience ( either 0 or 1 ) — наличие опыта исследовательской работы ( нет / есть)

```python
df_corr = df[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA', 'Research', 'Chance of Admit ']]
df_corr.corr()
```

```python
df_corr_0 = df_corr.loc[df_corr.Research == 0]
df_corr_0.corr(method='pearson')
```

```python
df_corr_1 = df_corr.loc[df_corr.Research == 1]
df_corr_1.corr(method='pearson')
```

```python
df_corr_OLS = df_corr_0[['TOEFL Score', 'CGPA']]
np.shape(df_corr_OLS)
```

```python
# Создаем вектор из единиц для коэффициента w_0 и записываем все векторы в СТОЛБЦЫ матрицы признаков А
TOEFL_Score=df_corr_OLS['TOEFL Score']
CGPA= df_corr_OLS['CGPA']
A= np.column_stack((np.ones(220), TOEFL_Score, CGPA))
A
```

```python
# Создаем целевой вектор
y = df_corr_0[['Chance of Admit ']] # объект типа dataframe - то, что нужно
```

```python
# вычислим OLS оценку для коэффициентов
w_hat=np.linalg.inv(A.T@A)@A.T@y
w_hat
```

```python
# прогноз
# добавились данные по новому городку:
TOEFL_Score_new = 107
CGPA_new=9.1
# делаем прогноз типичной соимости дома
Chance_new=round(w_hat.iloc[0], 3)+round(w_hat.iloc[1], 3)*TOEFL_Score_new+round(w_hat.iloc[2], 3)*CGPA_new
Chance_new
```

Стандартизация данных

```python
# метод .mean() позволяет вычислить арифметическое среднее значение вектора
meanTOEFL_Score = df_corr_OLS['TOEFL Score'].mean()
meanCGPA  = df_corr_OLS['CGPA'].mean()
mean_y = y.mean()
print('mean value of TOEFL_Score:', meanTOEFL_Score)
print('mean value of CGPA:', meanCGPA)
print('mean value of Chance of Admit:', mean_y)
```

```python
# Центрирование
TOEFL_Score_c = df_corr_OLS['TOEFL Score'] - meanTOEFL_Score
CGPA_c = df_corr_OLS['CGPA'] - meanCGPA
y_c = y - mean_y
```

```python
# вычисляем длины векторов для нормирования
TOEFL_Score_c_norm = np.linalg.norm(TOEFL_Score_c)
CGPA_c_norm = np.linalg.norm(CGPA_c)
y_c_norm = np.linalg.norm(y_c)
print('norm of TOEFL_Score:\n', TOEFL_Score_c_norm)
```

```python
# Нормирование: делим каждый центрированный вектор на его длину
TOEFL_Score_st=TOEFL_Score_c / TOEFL_Score_c_norm
CGPA_st=CGPA_c / CGPA_c_norm
y_st=y_c / y_c_norm
```

```python
# Матрица центрированных признаков - БЕЗ константы!
A_st=np.column_stack(( TOEFL_Score_st, CGPA_st))
A_st
```

```python
# OLS оценка коэффициентов центрированной регрессии
w_hat_st=np.linalg.inv(A_st.T@A_st)@A_st.T@y_st.values
w_hat_st
```
