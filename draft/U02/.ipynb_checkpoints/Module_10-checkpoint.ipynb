```python
'''Автоматически вывод на экран всех графиков и диаграмм без использования доп. ф-ций'''

%matplotlib inline
```

```python
df.plot()
```

При обращении к методу plot без указания параметров компьютер построит графики для всех имеющихся в датафреме количественных признаков.

```python
df.total_bill.plot(kind='hist')
```

```python
df.total_bill.plot(kind='hist', grid=True, title='Общая сумма счета')
```

```python
df['day'].value_counts().plot(kind='bar', grid=True, colormap='coolwarm', title='Кол-во посетителей по дням')
```

```python
#df['sex','smoker'].value_counts().plot(kind='bar', gride=True) #Error
df.groupby(by=['sex', 'smoker'])['tip'].mean().plot(kind='bar', grid=True)
df.pivot_table(index='sex', columns='smoker', values='tip').plot(kind='bar', grid=True)
```
